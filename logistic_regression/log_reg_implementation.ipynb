# Logistic Regression Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
# Create the dataset
x, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_classes=2,
    random_state=15,
)
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=15)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
# Model Training
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression()
```

```python
log_model.fit(x_train, y_train)
```

```python
y_pred = log_model.predict(x_test)
```

```python
y_pred
```

```python
log_model.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
print(f"Score: {score}\n")
cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{cm}\n")
print(classification_report(y_test, y_pred))
```

## Hyperparameter Tuning and Cross Validation

```python
model_log_opt = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 1.0, 0.1, 0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params = dict(penalty =  penalty, C =  c_values, solver =  solver)
```

```python
# GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid_model = GridSearchCV(estimator=model_log_opt, param_grid=params, scoring='accuracy', cv=cv, n_jobs = -1)
```

```python
grid_model
```

```python
grid_model.fit(x_train, y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
y_pred = grid_model.predict(x_test)
```

```python
score = accuracy_score(y_test, y_pred)
print(f"Score: {score}\n")
cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{cm}\n")
print(classification_report(y_test, y_pred))
```

## Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model_log_opt = LogisticRegression()
randomcv = RandomizedSearchCV(estimator=model_log_opt, param_distributions=params, cv=5, scoring='accuracy')
```

## Logistic Regression for Multiclass Classification Problem

```python
# Create the dataset
x, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_informative=3,
    n_classes=3,
    random_state=15,
)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(multi_class='ovr')
log_model.fit(x_train, y_train)
```

```python
y_pred = log_model.predict(x_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_test, y_pred)
print(f"Score: {score}\n")
cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{cm}\n")
print(classification_report(y_test, y_pred))
```
