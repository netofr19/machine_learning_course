# Algerian Forest Fires Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/Algerian_forest_fires_dataset_UPDATE.csv', header=1)
```

```python
df.head()
```

```python
df.info()
```

## Data Cleaning

```python
# Missing values
df[df.isna().any(axis=1)]
```

```python
df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
```

```python
df['Region'] = df['Region'].astype(int)
```

```python
df.head()
```

```python
# Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.iloc[[122]]
```

```python
# Remove the 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix spaces in columns names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

### Changes the required columns as integer data type

```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Changing the other columns to float datatype

```python
objects = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

## EDA and Feature Engineering

```python
# Drop day, month and year
df_copy = df.copy()
df_copy = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
# Encoding of the categoris in classes
df_copy['Classes'] = df_copy['Classes'].str.replace(' ', '')
df_copy['Classes'] = df_copy['Classes'].map({'notfire': 0, 'fire': 1})
```

```python
df_copy
```

```python
df_copy['Classes'].value_counts()
```

```python
# Plot density plot for all features
plt.style.use('seaborn-v0_8-notebook')
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Percentage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# Plotting piechart
classlabels = ['Fire', 'Not Fire']
plt.figure(figsize=(12,8))
plt.pie(percentage, labels=classlabels, autopct="%1.1f%%")
plt.title("Pie Chart of Classes")
plt.show()
```

```python
# Box Plots
sns.boxplot(data=df_copy, y='FWI', hue='Classes')
plt.show()
```

```python
# Monthly Fire Analysis
dftemp = df.loc[df['Region'] == 1]
dftemp['Classes'] = dftemp['Classes'].str.replace(' ', '')
dftemp['Classes'] = dftemp['Classes'].map({'notfire': 0, 'fire': 1})
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(data=dftemp, x='month', hue='Classes')
plt.ylabel('Number of fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title('Fire Analysis of Sidi-Bel Regions', weight='bold')
plt.show()
```

```python
# Monthly Fire Analysis
dftemp = df.loc[df['Region'] == 0]
dftemp['Classes'] = dftemp['Classes'].str.replace(' ', '')
dftemp['Classes'] = dftemp['Classes'].map({'notfire': 0, 'fire': 1})
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(data=dftemp, x='month', hue='Classes')
plt.ylabel('Number of fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title('Fire Analysis of Bejaia Regions', weight='bold')
plt.show()
```
