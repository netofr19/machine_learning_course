# 5 Number Summary and Box Plot

```python
import numpy as np
import random
```

```python
random.seed(101)
lst_marks = [random.randint(0, 101) for i in range(40)]
```

```python
mininum, q1, median, q3, maximum = np.quantile(lst_marks, [0, .25, .5, .75, 1])
```

```python
mininum, q1, median, q3, maximum
```

```python
iqr = q3 - q1
print(iqr)
```

```python
lower_fence = q1 - 1.5 * iqr
lower_fence
```

```python
higher_fence = q3 + 1.5 * iqr
higher_fence
```

```python
import seaborn as sns
```

```python
sns.boxplot(lst_marks)
```
